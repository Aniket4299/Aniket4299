```python
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn.impute import SimpleImputer
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error
```

```python
#read our data and store in a DataFrame
train_data = pd.read_csv('train.csv')
test_data = pd.read_csv('test.csv')
```

```python
train_data.head()
```

```python
#print a dataset
train_data.describe()
```

```python
#check size of dataset
train_data.shape
```

```python
#print all our features
features = train_data.columns
print("Features in the dataset:")
for feature in features:
    print(feature)
```

```python
#check for missing values
missing_val_count_by_column = (train_data.isnull().sum())
print(missing_val_count_by_column[missing_val_count_by_column > 0])
```

```python

sns.histplot(train_data.SalePrice, bins=40)
```

```python
#create target and feature variables
features = ['LotArea','1stFlrSF', '2ndFlrSF', 'FullBath', 'BedroomAbvGr', 'TotRmsAbvGrd']
y = train_data.SalePrice
X = train_data[features]
X.head()
```

```python
#data splitting
train_X, val_X, train_y, val_y = train_test_split(X, y, test_size=0.3, random_state=1)
```

```python
#dealing with missing values
my_imputer = SimpleImputer(strategy='mean')
train_X = pd.DataFrame(my_imputer.fit_transform(train_X))
val_X = pd.DataFrame(my_imputer.transform(val_X))
```

```python
#define our model
model = LinearRegression()
model.fit(train_X, train_y)
```

```python
#prediction
val_pred = model.predict(val_X)
val_mae = mean_absolute_error(val_pred, val_y)
print("Validation MAE for Linear Regression: {:,.0f}".format(val_mae))
```

```python
#plot a scatter
plt.scatter(val_y, val_pred)
```
